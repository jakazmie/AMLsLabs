```python
import os

image_path = '../../datasets'
input_images = os.path.join(image_path, 'aerial', 'train')
output_images = os.path.join(image_path, 'aerial-small')
```

```python
# Create folder structure in output
folders = os.listdir(input_images)
for folder in folders:
    os.makedirs(os.path.join(output_images, folder), exist_ok=True)
os.listdir(output_images)
```

```python
#Copy images
import random
import shutil

for folder in folders:
    filenames = os.listdir(os.path.join(input_images, folder))
    random.shuffle(filenames)
    for filename in filenames[0:1500]:
        shutil.copy(os.path.join(input_images, folder, filename), os.path.join(output_images, folder, filename))
        
    
    
```

```python
#Delete images

files = [os.path.join(output_images, folder, file) 
         for folder in os.listdir(output_images)
         for file in os.listdir(os.path.join(output_images, folder))]

for file in files:
    os.remove(file)
```
